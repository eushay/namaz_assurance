```python
import pandas as pd
import numpy as np

import plotly.graph_objects as go
from plotly.subplots import make_subplots
from plotly import offline
```

```python
category = 'Pants'
dir_id = 1
```

```python
df_accel = pd.read_csv(category +'/'+ str(dir_id) + '/Accelerometer.csv')
df_gyro = pd.read_csv(category +'/'+ str(dir_id) + '/Gyroscope.csv')
df_gravity = pd.read_csv(category +'/'+ str(dir_id) + '/Gravity.csv')
df_rot = pd.read_csv(category +'/'+ str(dir_id) + '/RotationVector.csv')
df_accelLin = pd.read_csv(category +'/'+ str(dir_id) + '/AccelerometerLinear.csv')
```

```python
df_gravity.head()
```

## Individual Plot

```python
df = df_rot
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter(x=df['Timestamp'], y=df['X'], mode='lines', name='Accel_X'))
fig.add_trace(go.Scatter(x=df['Timestamp'], y=df['Y'], mode='lines', name='Accel_Y'))
fig.add_trace(go.Scatter(x=df['Timestamp'], y=df['Z'], mode='lines', name='Accel_Z'))

fig.show()
```

## Combined Plot

```python
df1 = df_accel
df2 = df_accelLin
df3 = df_gyro
df4 = df_gravity
```

```python
fig = make_subplots(rows=2, cols=2, subplot_titles=("Accel", "AccelLinear", "Gyro", "Gravity"))

fig.add_trace(go.Scatter(x=df1['Timestamp'], y=df1['X'], mode='lines', name='X'),row=1, col=1)
fig.add_trace(go.Scatter(x=df1['Timestamp'], y=df1['Y'], mode='lines', name='Y'),row=1, col=1)
fig.add_trace(go.Scatter(x=df1['Timestamp'], y=df1['Z'], mode='lines', name='Z'),row=1, col=1)

fig.add_trace(go.Scatter(x=df2['Timestamp'], y=df2['X'], mode='lines', name='X'),row=1, col=2)
fig.add_trace(go.Scatter(x=df2['Timestamp'], y=df2['Y'], mode='lines', name='Y'),row=1, col=2)
fig.add_trace(go.Scatter(x=df2['Timestamp'], y=df2['Z'], mode='lines', name='Z'),row=1, col=2)

fig.add_trace(go.Scatter(x=df3['Timestamp'], y=df3['X'], mode='lines', name='X'),row=2, col=1)
fig.add_trace(go.Scatter(x=df3['Timestamp'], y=df3['Y'], mode='lines', name='Y'),row=2, col=1)
fig.add_trace(go.Scatter(x=df3['Timestamp'], y=df3['Z'], mode='lines', name='Z'),row=2, col=1)

fig.add_trace(go.Scatter(x=df4['Timestamp'], y=df4['X'], mode='lines', name='X'),row=2, col=2)
fig.add_trace(go.Scatter(x=df4['Timestamp'], y=df4['Y'], mode='lines', name='Y'),row=2, col=2)
fig.add_trace(go.Scatter(x=df4['Timestamp'], y=df4['Z'], mode='lines', name='Z'),row=2, col=2)

#fig.update_layout(height=600, width=800, title_text="Side By Side Subplots")


offline.plot(fig, filename = 'sensor_subplots.html')
#fig.show()  #  To show the plot in line
```
